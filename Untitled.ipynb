```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
df = pd.read_csv('books_data_clean.csv')
```

```python
# 2. Initial Inspection
print(df.head())
print(df.describe())
```

```python
# 3. Data Cleaning
# Fixing the negative publishing year
df = df[df['Publishing Year'] > 1900]
```

```python
# Dropping rows where the book name is missing
df.dropna(subset=['Book Name'], inplace=True)

```

```python
# 4. Exploratory Data Analysis & Visualization

# Distribution of Publishing Year
plt.figure(figsize=(10, 6))
plt.hist(df['Publishing Year'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Publishing Year')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()

```

```python
# Units Sold vs Sale Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sale price', y='units sold')
plt.title('Sale Price vs Units Sold')
plt.show()

```

```python
# Top 10 Authors by Gross Sales
top_authors = df.groupby('Author')['gross sales'].sum().sort_values(ascending=False).head(10)
top_authors.plot(kind='bar', figsize=(12, 6), color='salmon')
plt.title('Top 10 Authors by Total Gross Sales')
plt.ylabel('Gross Sales')
plt.show()

```

```python
# Language Distribution Pie Chart
lang_counts = df['language_code'].value_counts().head(5)
plt.figure(figsize=(10, 7))
plt.pie(lang_counts, labels=lang_counts.index, autopct='%1.1f%%', startangle=140 , pctdistance=0.95  )
plt.title('Top 5 Language Distribution')
plt.show()
```

```python
# Revenue by Publisher
publisher_rev = df.groupby('publisher')['publisher revenue'].sum().sort_values(ascending=False).head(10)
print("Top Publishers by Revenue:")
print(publisher_rev)
```
